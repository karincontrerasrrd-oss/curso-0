#Corte de Puerto Montt revoca resolución y ordena mantener el arresto domiciliario total de exalcalde
     
26-agosto-2025

####En fallo dividido, la Segunda Sala del tribunal de alzada –integrada por el ministro Moisés Montiel Torres, el fiscal judicial Danilo Báez Reyes y la abogada (i) María Paz Olavarría Pérez– revocó la resolución impugnada, dictada por el Juzgado de Garantía de Puerto Montt.


La Corte de Apelaciones de Puerto Montt acogió hoy -martes 26 de agosto- el recurso de apelación interpuesto por el municipio en calidad de querellante, con la adhesión del Consejo de Defensa del Estado, en contra de la resolución que sustituyó la medida cautelar de arresto domiciliario total por arresto domiciliario parcial y prohibición de comunicarse con los coimputados y funcionarios municipales, del exalcalde de la comuna Gervoy Paredes Rojas, imputado como autor de los delitos consumados de incremento patrimonial injustificado de funcionario público, cohecho, lavado de activos y fraude al fisco. Ilícitos perpetrados entre 2012 y 2023.

En fallo dividido (causa rol 1.005-2025), la Segunda Sala del tribunal de alzada –integrada por el ministro Moisés Montiel Torres, el fiscal judicial Danilo Báez Reyes y la abogada (i) María Paz Olavarría Pérez– revocó la resolución impugnada, dictada por el Juzgado de Garantía de Puerto Montt.

>“Que, del mérito de los antecedentes y de lo expuesto en la audiencia por los intervinientes, se desprende que, en este estadio procesal, existen elementos de convicción suficientes para estimar concurrentes los requisitos de las letras a) y b) del artículo 140 del Código Procesal Penal, esto es, la existencia material de los hechos investigados y aquellos tendientes a presumir fundadamente la participación que en ellos se atribuye al imputado, lo cual además, no fue controvertido por la defensa”, consigna el fallo.

La resolución agrega:

>> “Que, respecto de la necesidad de cautela prevista en la letra c) del mismo artículo, se advierte que la libertad del imputado constituye un peligro para la seguridad de la sociedad, dado el número, naturaleza y gravedad de los ilícitos, la forma de comisión, el alto perjuicio ocasionado al patrimonio público, la pena de crimen asignada al delito de fraude al fisco, la prognosis de pena y la eventual forma de cumplimiento en caso de condena”.

>>“Además, los nuevos antecedentes que dan cuenta de la existencia de deudas del imputado y el transcurso del tiempo no constituyen una variación sustancial en las circunstancias tomadas en consideración al momento de decretarse el arresto domiciliario total (…)”.

>>Por tanto, se resuelve que: “se revoca la resolución de diecinueve de agosto del año en curso, dictada por don Rolando Díaz Coloma, juez del Juzgado de Garantía de Puerto Montt, y en su lugar, se declara que se mantiene la medida cautelar de arresto domiciliario total y el arraigo nacional respecto del imputado Gervoy Amador Paredes Rojas”.

Decisión acordada con el voto en contra del fiscal judicial Danilo Báez Reyes, quien estuvo por confirmar la resolución en alzada.
